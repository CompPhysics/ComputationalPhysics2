## 1  Physical Model and Trial Wave Function

### 1.1  The Hamiltonian

We study $N$ **non-interacting spinless bosons** confined in a $d$-dimensional isotropic harmonic oscillator trap.  In natural units ($\hbar = m = \omega = 1$) the Hamiltonian is

$$H = \sum_{i=1}^{N}\left(-\frac{1}{2}\nabla_i^2 + \frac{1}{2}r_i^2\right),$$

where $r_i = |\mathbf{r}_i|$ and $\nabla_i^2$ is the Laplacian acting on particle $i$.

The exact ground-state energy is

$$E_0 = \frac{Nd}{2},$$

which follows immediately from the single-particle ground-state energy $d/2$ summed over all $N$ bosons.

### 1.2  Trial wave function

We use a **Gaussian ansatz** with a single variational parameter $\alpha > 0$:

$$\Psi_\alpha(\mathbf{r}_1,\ldots,\mathbf{r}_N) = \prod_{i=1}^{N} e^{-\alpha r_i^2} = \exp\!\left(-\alpha\sum_{i=1}^{N} r_i^2\right).$$

This is the exact ground state when $\alpha = 1/2$, so the VMC optimiser should converge to $\alpha^* = 0.5$ starting from any positive initial value.

The exact energy as a function of $\alpha$ is available analytically:

$$E(\alpha) = \frac{Nd}{2}\left(\alpha + \frac{1}{4\alpha}\right),$$

which has a minimum $E(1/2) = Nd/2$ at $\alpha=1/2$. This serves as a **benchmark** for the optimiser.

### 1.3  The logarithmic derivative

The single variational parameter means $\boldsymbol{\alpha} = (\alpha)$ and there is one logarithmic derivative operator:

$$O_1(\mathbf{x}) = \frac{\partial\ln\Psi_\alpha}{\partial\alpha} = -\sum_{i=1}^{N} r_i^2.$$

The local energy (kinetic + potential, computed via the Laplacian) is

$$E_\mathrm{loc}(\mathbf{x}) = \frac{H\Psi_\alpha}{\Psi_\alpha} = Nd\alpha + \left(\frac{1}{2} - 2\alpha^2\right)\sum_{i=1}^N r_i^2.$$

## 2  Imaginary-Time Evolution and Variational Projection

### 2.1  Imaginary-time evolution

The **exact** imaginary-time propagation of a state is

$$|\Psi(\tau)\rangle = e^{-\tau H}|\Psi(0)\rangle.$$

Expanding in energy eigenstates $H|n\rangle = E_n|n\rangle$:

$$|\Psi(\tau)\rangle = \sum_n c_n\, e^{-\tau E_n}|n\rangle.$$

For large $\tau$, provided $c_0 \neq 0$, the ground state dominates:

$$|\Psi(\tau)\rangle \propto e^{-\tau E_0}|0\rangle.$$

Imaginary-time propagation **filters out the ground state**.

### 2.2  Variational projection

In VMC we restrict to the variational manifold

$$\mathcal{M} = \left\{|\Psi(\boldsymbol{\alpha})\rangle \;\middle|\; \boldsymbol{\alpha}\in\mathbb{R}^p\right\}.$$

An exact step $(1-\delta\tau\, H)|\Psi(\boldsymbol{\alpha})\rangle$ **does not remain in $\mathcal{M}$** for a general parametric family.  The **stochastic reconfiguration** idea is:

1. Apply one infinitesimal imaginary-time step.
2. Project the result back onto $\mathcal{M}$.
3. Read off the implied parameter update $\boldsymbol{\alpha}\to\boldsymbol{\alpha}+\delta\boldsymbol{\alpha}$.

A norm-preserving form of the infinitesimal step is

$$|\Psi_\mathrm{exact}'\rangle \approx \left[1-\delta\tau\bigl(H-E(\boldsymbol{\alpha})\bigr)\right]|\Psi(\boldsymbol{\alpha})\rangle,$$

where subtracting $E(\boldsymbol{\alpha})$ removes the trivial component parallel to $|\Psi\rangle$ (it is orthogonal to $|\Psi\rangle$ to first order in $\delta\tau$, ensuring the update lies in the tangent space of projective Hilbert space).


---
## 3  Tangent-Space Expansion and the Projection Principle

### 3.1  Taylor expansion of the variational state

A first-order Taylor expansion of the variational state gives

$$|\Psi(\boldsymbol{\alpha}+\delta\boldsymbol{\alpha})\rangle = |\Psi\rangle + \sum_k \delta\alpha_k |\Psi_k\rangle + \mathcal{O}(\delta\alpha^2),$$

where the **tangent vectors** are

$$|\Psi_k\rangle \equiv \frac{\partial}{\partial\alpha_k}|\Psi(\boldsymbol{\alpha})\rangle.$$

### 3.2  Minimising the mismatch

Define the **mismatch vector**

$$|\Delta\rangle = |\Psi_\mathrm{var}'\rangle - |\Psi_\mathrm{exact}'\rangle = \sum_k \delta\alpha_k|\Psi_k\rangle + \delta\tau(H-E)|\Psi\rangle.$$

The SR update $\delta\boldsymbol{\alpha}$ minimises the squared norm

$$\mathcal{D} = \langle\Delta|\Delta\rangle.$$

Expanding and retaining terms of order $\delta\alpha\cdot\delta\tau$:

$$\mathcal{D} = \sum_{ij}\delta\alpha_i\delta\alpha_j\langle\Psi_i|\Psi_j\rangle + 2\delta\tau\sum_i\delta\alpha_i\,\mathrm{Re}\langle\Psi_i|(H-E)|\Psi\rangle + \mathcal{O}(\delta\tau^2).$$

### 3.3  Stationarity condition

Setting $\partial\mathcal{D}/\partial(\delta\alpha_i)=0$ yields the **projected imaginary-time equation** in the tangent basis:

$$\sum_j \langle\Psi_i|\Psi_j\rangle\,\delta\alpha_j = -\delta\tau\,\mathrm{Re}\langle\Psi_i|(H-E)|\Psi\rangle.$$


---
## 4  Derivation of the SR Linear System

### 4.1  Logarithmic derivative operators

Let $x$ denote a many-body configuration and write the wave function as $\Psi(x)$.  Define the **logarithmic derivative operators**

$$O_k(x) = \frac{1}{\Psi(x)}\frac{\partial\Psi(x)}{\partial\alpha_k} = \frac{\partial\ln\Psi(x)}{\partial\alpha_k}.$$

Monte Carlo samples configurations with probability

$$P(x) = \frac{|\Psi(x)|^2}{\langle\Psi|\Psi\rangle},$$

so expectation values are estimated as sample means: $\langle A\rangle \approx \frac{1}{N_s}\sum_{\ell=1}^{N_s} A(x^{(\ell)})$.

### 4.2  The metric matrix $S$

Working with the normalised state $|\overline{\Psi}\rangle = |\Psi\rangle/\sqrt{\langle\Psi|\Psi\rangle}$ and its tangent vectors, one finds

$$S_{ij} = \langle O_i O_j\rangle - \langle O_i\rangle\langle O_j\rangle.$$

This is the **covariance matrix of the logarithmic derivatives**.  It is symmetric positive semidefinite, and defines a Riemannian metric on $\mathcal{M}$:

$$ds^2 = \sum_{ij} S_{ij}\,d\alpha_i\,d\alpha_j.$$

### 4.3  The generalised force $\boldsymbol{g}$

The local energy is

$$E_\mathrm{loc}(x) = \frac{(H\Psi)(x)}{\Psi(x)}.$$

The right-hand side of the stationarity condition becomes

$$g_i = \langle O_i E_\mathrm{loc}\rangle - \langle O_i\rangle\langle E_\mathrm{loc}\rangle,$$

the covariance of the logarithmic derivative with the local energy.

### 4.4  The SR linear system

Collecting everything:

$$\boxed{S\,\delta\boldsymbol{\alpha} = -\delta\tau\,\boldsymbol{g}, \qquad \delta\boldsymbol{\alpha} = -\delta\tau\,S^{-1}\boldsymbol{g}.}$$

**Comparison with gradient descent.** Ordinary gradient descent updates $\delta\boldsymbol{\alpha}_\mathrm{GD} = -\eta\boldsymbol{g}$, while SR uses $\delta\boldsymbol{\alpha}_\mathrm{SR} = -\eta\,S^{-1}\boldsymbol{g}$.  The inverse metric $S^{-1}$ rescales parameter directions according to the geometry of the wave-function manifold, giving better conditioning and a direct connection to imaginary-time evolution.


---
## 5  Natural Gradient, Regularisation, and the SR Algorithm

### 5.1  SR as the natural gradient

Because $S$ is the Riemannian metric on $\mathcal{M}$, the steepest-descent direction in that metric is

$$\delta\boldsymbol{\alpha} \propto -S^{-1}\boldsymbol{g},$$

which is precisely the **natural gradient** update (Amari 1998).  SR therefore inherits all the favourable properties of the natural gradient:

- Reduced sensitivity to redundant parameterisations.
- Better conditioning than plain gradient descent.
- Covariance under parameter reparameterisation.

### 5.2  Regularisation

In practice $S$ is estimated from a finite Monte Carlo sample and may be singular or ill-conditioned.  The standard fix is **diagonal regularisation**:

$$S \to S + \lambda I, \qquad \lambda > 0.$$

The regularised system

$$(S+\lambda I)\,\delta\boldsymbol{\alpha} = -\delta\tau\,\boldsymbol{g}$$

- stabilises the matrix inversion,
- damps directions with negligible Fisher information,
- interpolates between SR ($\lambda\to 0$) and plain gradient descent ($\lambda\to\infty$).

### 5.3  SR algorithm summary

At each iteration:

1. Sample $N_s$ configurations $x^{(\ell)}$ from $P(x)=|\Psi(x)|^2/\langle\Psi|\Psi\rangle$ using Metropolis–Hastings.
2. Compute logarithmic derivatives $O_k(x^{(\ell)})$ and local energies $E_\mathrm{loc}(x^{(\ell)})$ at each configuration.
3. Estimate
$$S_{ij} = \langle O_i O_j\rangle - \langle O_i\rangle\langle O_j\rangle, \qquad g_i = \langle O_i E_\mathrm{loc}\rangle - \langle O_i\rangle\langle E_\mathrm{loc}\rangle.$$
4. Solve $(S+\lambda I)\,\delta\boldsymbol{\alpha} = -\delta\tau\,\boldsymbol{g}$.
5. Update $\boldsymbol{\alpha} \leftarrow \boldsymbol{\alpha} + \delta\boldsymbol{\alpha}$.

## 6  Python Implementation

We now implement the full VMC + SR algorithm in pure NumPy.  Every quantity that appears in the theory above — the metric matrix $S$, the force vector $\boldsymbol{g}$, the Metropolis sampler, and the local energy — is computed explicitly so that the code mirrors the mathematics one-to-one.


```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.gridspec as gridspec
from scipy.linalg import solve
import warnings
warnings.filterwarnings('ignore')

# Reproducibility
np.random.seed(42)

plt.rcParams.update({
    'figure.dpi': 120,
    'font.size': 12,
    'axes.labelsize': 13,
    'axes.titlesize': 14,
    'lines.linewidth': 2.0,
    'axes.grid': True,
    'grid.alpha': 0.35,
})
print(f'NumPy {np.__version__}  ready.')
```

### 6.1  Physical model

The trial wave function and all derived quantities are defined here.  The code is written for $N$ particles in $d$ dimensions with a single variational parameter $\alpha$.

```python
# ── Physical parameters ───────────────────────────────────────────────────────
N_PARTICLES = 5    # number of non-interacting bosons
DIM         = 3    # spatial dimension
EXACT_E0    = N_PARTICLES * DIM / 2.0   # exact ground-state energy Nd/2
EXACT_ALPHA = 0.5                        # variational minimum


# ── Trial wave function  Ψ_α(x) = exp(-α Σ r_i^2) ───────────────────────────
def log_psi(x, alpha):
    """Log of the trial wave function (unnormalised).

    Parameters
    ----------
    x     : ndarray (N, d)  -- particle positions
    alpha : float           -- variational parameter

    Returns
    -------
    float : ln Psi_alpha(x) = -alpha * sum_i |r_i|^2
    """
    return -alpha * np.sum(x**2)


def psi(x, alpha):
    """Trial wave function value."""
    return np.exp(log_psi(x, alpha))


# ── Local energy  E_loc(x) = (H Psi)(x) / Psi(x) ────────────────────────────
def local_energy(x, alpha):
    """Analytic local energy for the Gaussian ansatz.

    For Psi = exp(-alpha * sum r_i^2):
        Kinetic: -1/2 nabla^2 Psi / Psi = alpha*d*N - 2*alpha^2 * sum r_i^2
        Potential: 1/2 * sum r_i^2

    So  E_loc = N*d*alpha + (1/2 - 2*alpha^2) * sum r_i^2
    """
    r2_total = np.sum(x**2)          # sum_i |r_i|^2
    N, d     = x.shape
    return N * d * alpha + (0.5 - 2.0 * alpha**2) * r2_total


# ── Logarithmic derivative  O_1(x) = d ln Psi / d alpha ──────────────────────
def log_deriv(x):
    """Logarithmic derivative O_1(x) = d ln Psi_alpha / d alpha = -sum r_i^2.

    (Independent of alpha for this ansatz.)
    """
    return -np.sum(x**2)   # scalar for the single-parameter case


# ── Exact analytic energy  E(alpha) = N*d/2 * (alpha + 1/(4*alpha)) ──────────
def exact_energy(alpha, N=N_PARTICLES, d=DIM):
    """Analytic expectation value of H for the Gaussian trial state."""
    return N * d / 2.0 * (alpha + 1.0 / (4.0 * alpha))


# Quick sanity check
x_test = np.zeros((N_PARTICLES, DIM))
print(f'N = {N_PARTICLES} bosons, d = {DIM} dimensions')
print(f'Exact ground-state energy     E_0     = {EXACT_E0:.4f}')
print(f'Exact minimum of E(alpha)     alpha*  = {EXACT_ALPHA:.4f}')
print(f'E_loc(0, alpha=0.5)           = {local_energy(x_test, 0.5):.4f}  (should be 0)')
print(f'E_loc(0, alpha=1.0)           = {local_energy(x_test, 1.0):.4f}  (should be 0)')
print(f'Analytic E(alpha=0.5)         = {exact_energy(0.5):.4f}  (should be {EXACT_E0:.4f})')
print(f'Analytic E(alpha=1.0)         = {exact_energy(1.0):.4f}')
```

### 6.2  Metropolis–Hastings sampler

We sample configurations $x$ from

$$P(x) = \frac{|\Psi_\alpha(x)|^2}{\langle\Psi_\alpha|\Psi_\alpha\rangle}$$

using the **Metropolis–Hastings algorithm** with a symmetric Gaussian proposal $x' = x + \delta\,\xi$, $\xi\sim\mathcal{N}(0,1)$.  The acceptance ratio simplifies to

$$A(x'|x) = \min\!\left(1,\frac{|\Psi(x')|^2}{|\Psi(x)|^2}\right) = \min\!\left(1, e^{2[\ln\Psi(x')-\ln\Psi(x)]}\right).$$

The step size $\delta$ is tuned to achieve an acceptance rate near 50%.

```python
def metropolis_sampler(alpha, N_samples, step_size=0.5,
                       N_equil=2000, N_particles=N_PARTICLES, dim=DIM):
    """Metropolis-Hastings sampler for |Psi_alpha(x)|^2.

    Parameters
    ----------
    alpha       : float   -- variational parameter
    N_samples   : int     -- number of samples to collect after equilibration
    step_size   : float   -- Gaussian proposal width delta
    N_equil     : int     -- equilibration steps (discarded)
    N_particles : int     -- number of particles
    dim         : int     -- spatial dimension

    Returns
    -------
    samples     : ndarray (N_samples, N_particles, dim)
    accept_rate : float   -- fraction of accepted moves
    """
    # Initialise at random positions
    x_curr    = np.random.randn(N_particles, dim) * 0.5
    lp_curr   = log_psi(x_curr, alpha)   # log |Psi(x)|  (not squared; we compare 2*diff)

    samples   = np.zeros((N_samples, N_particles, dim))
    n_accept  = 0
    n_total   = N_equil + N_samples

    for step in range(n_total):
        # Gaussian proposal: move all particles simultaneously
        x_prop  = x_curr + step_size * np.random.randn(N_particles, dim)
        lp_prop = log_psi(x_prop, alpha)

        # Metropolis acceptance:  accept with prob min(1, |Psi(x')|^2 / |Psi(x)|^2)
        # = min(1, exp(2*(lp_prop - lp_curr)))
        log_ratio = 2.0 * (lp_prop - lp_curr)
        if np.log(np.random.uniform()) < log_ratio:
            x_curr  = x_prop
            lp_curr = lp_prop
            if step >= N_equil:
                n_accept += 1

        if step >= N_equil:
            samples[step - N_equil] = x_curr

    accept_rate = n_accept / N_samples
    return samples, accept_rate


# Test the sampler at the exact minimum
test_samples, test_acc = metropolis_sampler(0.5, N_samples=5000, step_size=0.5)
E_test = np.mean([local_energy(s, 0.5) for s in test_samples])
print(f'Sampler test at alpha=0.5:')
print(f'  Acceptance rate  : {test_acc:.3f}  (target ~0.50)')
print(f'  <E_loc>          : {E_test:.5f}  (exact: {EXACT_E0:.5f})')
```

### 6.3  VMC energy and gradient estimator

Given samples from $P(x)$ we estimate:

$$\langle E\rangle \approx \frac{1}{N_s}\sum_{\ell=1}^{N_s} E_\mathrm{loc}(x^{(\ell)}),$$

$$S_{ij} \approx \frac{1}{N_s}\sum_\ell O_i(x^{(\ell)})O_j(x^{(\ell)}) - \bar{O}_i\bar{O}_j,$$

$$g_i \approx \frac{1}{N_s}\sum_\ell O_i(x^{(\ell)})E_\mathrm{loc}(x^{(\ell)}) - \bar{O}_i\bar{E},$$

where $\bar{O}_i = \frac{1}{N_s}\sum_\ell O_i(x^{(\ell)})$ and $\bar{E} = \frac{1}{N_s}\sum_\ell E_\mathrm{loc}(x^{(\ell)})$.

For the single-parameter Gaussian ansatz, $S$ and $\boldsymbol{g}$ reduce to scalar quantities:

$$S_{11} = \langle O_1^2\rangle - \langle O_1\rangle^2, \qquad g_1 = \langle O_1 E_\mathrm{loc}\rangle - \langle O_1\rangle\langle E_\mathrm{loc}\rangle.$$

```python
def vmc_estimate(alpha, N_samples=4000, step_size=0.5):
    """VMC energy and SR ingredients for a given alpha.

    Returns
    -------
    E_mean   : float  -- <E_loc>
    E_std    : float  -- standard error on <E_loc>
    S_mat    : float  -- S matrix (1x1 for single parameter)
    g_vec    : float  -- force vector g (scalar here)
    acc_rate : float  -- Metropolis acceptance rate
    """
    samples, acc_rate = metropolis_sampler(alpha, N_samples,
                                           step_size=step_size)
    Ns = len(samples)

    # Evaluate O_k and E_loc at each sampled configuration
    O  = np.array([log_deriv(s)    for s in samples])   # shape (Ns,)
    EL = np.array([local_energy(s, alpha) for s in samples])  # shape (Ns,)

    # Mean values
    O_mean  = O.mean()    # <O_1>
    E_mean  = EL.mean()   # <E_loc>
    E_std   = EL.std() / np.sqrt(Ns)   # standard error

    # Metric matrix  S_11 = <O^2> - <O>^2  (covariance of log-derivatives)
    S_mat = (O**2).mean() - O_mean**2

    # Force vector  g_1 = <O * E_loc> - <O> * <E_loc>  (covariance with local energy)
    g_vec = (O * EL).mean() - O_mean * E_mean

    return E_mean, E_std, S_mat, g_vec, acc_rate


# Scan alpha to verify analytic energy curve
alphas_scan = np.linspace(0.2, 1.2, 30)
E_vmc   = []
E_exact = [exact_energy(a) for a in alphas_scan]

for a in alphas_scan:
    E_m, _, _, _, _ = vmc_estimate(a, N_samples=3000)
    E_vmc.append(E_m)

fig, ax = plt.subplots(figsize=(8, 5))
ax.plot(alphas_scan, E_exact, color='steelblue', lw=2.5, label='Analytic $E(\\alpha)$')
ax.scatter(alphas_scan, E_vmc,  color='tomato',    s=40,   label='VMC estimate', zorder=5)
ax.axvline(EXACT_ALPHA, color='k', ls=':', lw=1.5,
           label=rf'$\alpha^*={EXACT_ALPHA}$')
ax.axhline(EXACT_E0,    color='grey', ls='-.', lw=1.2,
           label=rf'$E_0={EXACT_E0:.1f}$')
ax.set_xlabel(r'Variational parameter $\alpha$')
ax.set_ylabel(r'Energy $\langle E \rangle$')
ax.set_title(rf'VMC energy vs $\alpha$: $N={N_PARTICLES}$ bosons, $d={DIM}$')
ax.legend(fontsize=10)
plt.tight_layout(); plt.show()
print('VMC energy curve matches analytic result.')
```

### 6.4  Stochastic Reconfiguration optimiser

The SR update for the single-parameter case reduces to

$$\delta\alpha = -\delta\tau\,\frac{g_1}{S_{11}+\lambda},$$

which is just a rescaled gradient step with a geometry-aware learning rate $1/(S_{11}+\lambda)$.  For multiple parameters it generalises to the full linear system $(S+\lambda I)\,\delta\boldsymbol{\alpha}=-\delta\tau\,\boldsymbol{g}$, which is solved with `scipy.linalg.solve`.

We also implement plain **gradient descent** (GD) for comparison:

$$\delta\alpha_\mathrm{GD} = -\eta\,g_1.$$

```python
def sr_optimizer(
    alpha_init,
    n_iterations  = 200,
    dtau          = 0.05,    # SR step size  delta_tau
    lam           = 1e-3,    # SR regularisation  lambda
    N_samples     = 3000,    # MC samples per iteration
    step_size     = 0.5,     # Metropolis step size
    verbose       = True,
):
    """Stochastic Reconfiguration optimiser for a single variational parameter.

    At each iteration it:
    1. Runs a Metropolis-Hastings Markov chain to sample |Psi_alpha|^2.
    2. Estimates S_11 (covariance of log-derivatives) and g_1 (covariance
       of log-derivative with local energy) from the samples.
    3. Solves (S_11 + lambda) * d_alpha = -dtau * g_1.
    4. Updates alpha <- alpha + d_alpha.

    Parameters
    ----------
    alpha_init    : float  -- starting value of alpha
    n_iterations  : int    -- number of SR steps
    dtau          : float  -- imaginary-time step (SR learning rate)
    lam           : float  -- diagonal regularisation for S
    N_samples     : int    -- Metropolis samples per iteration
    step_size     : float  -- Metropolis proposal width
    verbose       : bool   -- print progress every 20 steps

    Returns
    -------
    history : dict with keys 'alpha', 'energy', 'energy_std', 'S', 'g',
              'dalpha', 'accept_rate'
    """
    alpha   = float(alpha_init)
    history = {
        'alpha'      : [],
        'energy'     : [],
        'energy_std' : [],
        'S'          : [],
        'g'          : [],
        'dalpha'     : [],
        'accept_rate': [],
    }

    for it in range(n_iterations):
        # ── Step 1–3: VMC estimation ──────────────────────────────────────────
        E_mean, E_std, S_11, g_1, acc = vmc_estimate(
            alpha, N_samples=N_samples, step_size=step_size)

        # ── Step 4: SR update  d_alpha = -dtau * g_1 / (S_11 + lambda) ──────
        # (Generalises to:  (S + lambda*I) d_alpha = -dtau * g  for p>1)
        d_alpha = -dtau * g_1 / (S_11 + lam)

        # ── Step 5: parameter update ──────────────────────────────────────────
        alpha = max(alpha + d_alpha, 1e-4)   # keep alpha > 0

        # Record
        history['alpha']      .append(alpha)
        history['energy']     .append(E_mean)
        history['energy_std'] .append(E_std)
        history['S']          .append(S_11)
        history['g']          .append(g_1)
        history['dalpha']     .append(d_alpha)
        history['accept_rate'].append(acc)

        if verbose and (it % 20 == 0 or it == n_iterations - 1):
            print(f'  Iter {it+1:4d}/{n_iterations}  '
                  f'alpha={alpha:.6f}  '
                  f'E={E_mean:.5f} +/- {E_std:.5f}  '
                  f'S={S_11:.4e}  g={g_1:.4e}  '
                  f'acc={acc:.3f}')

    return {k: np.array(v) for k, v in history.items()}


print('SR optimiser defined. Running optimisation...')
```

```python
def gd_optimizer(
    alpha_init,
    n_iterations = 200,
    eta          = 0.05,    # GD learning rate
    N_samples    = 3000,
    step_size    = 0.5,
    verbose      = True,
):
    """Plain gradient descent:  delta_alpha = -eta * g_1.

    Identical setup to sr_optimizer but without the S-matrix rescaling.
    """
    alpha   = float(alpha_init)
    history = {
        'alpha'      : [],
        'energy'     : [],
        'energy_std' : [],
        'g'          : [],
    }

    for it in range(n_iterations):
        E_mean, E_std, S_11, g_1, acc = vmc_estimate(
            alpha, N_samples=N_samples, step_size=step_size)

        # Plain gradient step (no S^{-1} rescaling)
        d_alpha = -eta * g_1
        alpha   = max(alpha + d_alpha, 1e-4)

        history['alpha']      .append(alpha)
        history['energy']     .append(E_mean)
        history['energy_std'] .append(E_std)
        history['g']          .append(g_1)

        if verbose and (it % 20 == 0 or it == n_iterations - 1):
            print(f'  Iter {it+1:4d}/{n_iterations}  '
                  f'alpha={alpha:.6f}  '
                  f'E={E_mean:.5f} +/- {E_std:.5f}')

    return {k: np.array(v) for k, v in history.items()}

print('GD optimiser defined.')
```

### 6.5  Running the SR optimiser

We start from $\alpha_0 = 0.8$ (displaced from the exact minimum $\alpha^*=0.5$) and run 150 SR iterations.

```python
ALPHA_INIT   = 0.8    # starting point (far from alpha* = 0.5)
N_ITER       = 150
N_SAMPLES    = 4000

print('=' * 65)
print('  Stochastic Reconfiguration optimisation')
print(f'  Starting alpha = {ALPHA_INIT},  target alpha* = {EXACT_ALPHA}')
print('=' * 65)
hist_sr = sr_optimizer(
    alpha_init   = ALPHA_INIT,
    n_iterations = N_ITER,
    dtau         = 0.05,
    lam          = 1e-3,
    N_samples    = N_SAMPLES,
    step_size    = 0.5,
    verbose      = True,
)
```

```python
print('=' * 65)
print('  Gradient descent optimisation  (for comparison)')
print(f'  Starting alpha = {ALPHA_INIT}')
print('=' * 65)
hist_gd = gd_optimizer(
    alpha_init   = ALPHA_INIT,
    n_iterations = N_ITER,
    eta          = 0.05,
    N_samples    = N_SAMPLES,
    step_size    = 0.5,
    verbose      = True,
)
```

### 7.3  Multi-parameter extension: the full $S$ matrix

To illustrate the full SR linear system for $p > 1$ parameters, we extend the ansatz to a **two-parameter Gaussian**:

$$\Psi_{\alpha_1,\alpha_2}(\mathbf{r}_1,\ldots,\mathbf{r}_N) = \exp\!\left(-\alpha_1\sum_{i=1}^{N}r_i^2 - \alpha_2\sum_{i=1}^{N}r_i^4\right).$$

For this ansatz the logarithmic derivatives are

$$O_1(x) = -\sum_i r_i^2, \qquad O_2(x) = -\sum_i r_i^4,$$

and the metric matrix is

$$S = \begin{pmatrix}
\langle O_1^2\rangle-\langle O_1\rangle^2 & 
\langle O_1 O_2\rangle-\langle O_1\rangle\langle O_2\rangle \\
\langle O_1 O_2\rangle-\langle O_1\rangle\langle O_2\rangle & 
\langle O_2^2\rangle-\langle O_2\rangle^2
\end{pmatrix}.$$

The exact minimum is $\boldsymbol{\alpha}^* = (0.5, 0)$ (the $r^4$ term vanishes in the exact ground state).  This example demonstrates how the $2\times 2$ SR linear system is assembled and solved, and verifies that $S$ is indeed positive semidefinite.

```python
# ── Two-parameter ansatz: Psi = exp(-a1*sum r^2 - a2*sum r^4) ───────────────

def log_psi_2p(x, alpha):
    """Log-wave-function for the two-parameter ansatz."""
    a1, a2 = alpha
    r2 = np.sum(x**2)
    r4 = np.sum(x**4)
    return -a1 * r2 - a2 * r4


def local_energy_2p(x, alpha):
    """Local energy for the two-parameter ansatz (via finite differences)."""
    a1, a2  = alpha
    N, d    = x.shape
    r2      = np.sum(x**2)
    r4      = np.sum(x**4)
    # Analytic: kinetic energy from -1/2 nabla^2 psi / psi
    # For phi = -a1 r^2 - a2 r^4:
    # nabla_i^2 phi = -2*a1*d - 4*a2*(d+2)*r_i^2
    # (nabla_i phi)^2 = (2*a1*r_i + 4*a2*r_i^3)^2
    # E_kin = sum_i [ -1/2 * nabla_i^2 phi - 1/2 * (nabla_i phi)^2 ]
    kinetic = 0.0
    for i in range(N):
        ri2       = np.sum(x[i]**2)
        lap_phi   = -2*a1*d - 4*a2*(d + 2)*ri2
        grad_phi2 = np.sum((2*a1*x[i] + 4*a2*x[i]**3)**2)
        kinetic  += -0.5 * lap_phi - 0.5 * grad_phi2
    potential = 0.5 * r2
    return kinetic + potential


def log_deriv_2p(x):
    """Logarithmic derivatives [O_1, O_2] for the two-parameter ansatz."""
    return np.array([-np.sum(x**2), -np.sum(x**4)])   # shape (2,)


def sr_step_2p(alpha, N_samples=3000, dtau=0.05, lam=1e-3,
               step_size=0.5):
    """One SR step for the two-parameter ansatz.

    Returns updated alpha, energy, S matrix (2x2), g vector (2,).
    """
    # Sample from |Psi|^2 via Metropolis
    x_curr = np.random.randn(N_PARTICLES, DIM) * 0.5
    lp_cur = log_psi_2p(x_curr, alpha)
    samples = []
    for _ in range(2000 + N_samples):    # 2000 equilibration
        x_prop = x_curr + step_size * np.random.randn(N_PARTICLES, DIM)
        lp_new = log_psi_2p(x_prop, alpha)
        if np.log(np.random.uniform()) < 2*(lp_new - lp_cur):
            x_curr, lp_cur = x_prop, lp_new
        if len(samples) < N_samples:
            samples.append(x_curr.copy())
        else:
            pass  # done after equilibration
    # correct: collect only after equil
    x_curr = np.random.randn(N_PARTICLES, DIM) * 0.5
    lp_cur = log_psi_2p(x_curr, alpha)
    configs = []
    for step in range(2000 + N_samples):
        x_prop = x_curr + step_size * np.random.randn(N_PARTICLES, DIM)
        lp_new = log_psi_2p(x_prop, alpha)
        if np.log(np.random.uniform()) < 2*(lp_new - lp_cur):
            x_curr, lp_cur = x_prop, lp_new
        if step >= 2000:
            configs.append(x_curr.copy())

    # Vectorise
    O_all  = np.array([log_deriv_2p(c) for c in configs])   # (Ns, 2)
    EL_all = np.array([local_energy_2p(c, alpha) for c in configs])  # (Ns,)

    O_mean  = O_all.mean(axis=0)    # (2,)
    E_mean  = EL_all.mean()

    # S matrix (2x2):  S_ij = <O_i O_j> - <O_i><O_j>
    S_mat = (O_all[:, :, None] * O_all[:, None, :]).mean(axis=0) \
            - np.outer(O_mean, O_mean)   # (2, 2)

    # g vector (2,):  g_i = <O_i E_loc> - <O_i><E_loc>
    g_vec = (O_all * EL_all[:, None]).mean(axis=0) - O_mean * E_mean  # (2,)

    # Solve (S + lam I) d_alpha = -dtau g
    S_reg   = S_mat + lam * np.eye(2)
    d_alpha = solve(S_reg, -dtau * g_vec)
    alpha_new = alpha + d_alpha
    alpha_new[0] = max(alpha_new[0], 1e-4)   # keep a1 > 0

    return alpha_new, E_mean, S_mat, g_vec


# ── Run 2-parameter SR ────────────────────────────────────────────────────────
alpha2p  = np.array([0.8, 0.1])   # start away from (0.5, 0)
traj_2p  = []

print('Running 2-parameter SR  (alpha* = [0.5, 0.0]) ...')
for it in range(80):
    alpha2p, E_m, S_m, g_m = sr_step_2p(alpha2p, N_samples=3000)
    traj_2p.append((alpha2p.copy(), E_m, np.linalg.eigvalsh(S_m)))
    if it % 20 == 0 or it == 79:
        print(f'  iter {it+1:3d}: alpha={alpha2p},  E={E_m:.5f}')

alphas_2p = np.array([t[0] for t in traj_2p])
energies_2p = np.array([t[1] for t in traj_2p])
eigenvalues_2p = np.array([t[2] for t in traj_2p])

fig, axes = plt.subplots(1, 3, figsize=(15, 5))

ax = axes[0]
ax.plot(alphas_2p[:, 0], color='steelblue', lw=2, label=r'$\alpha_1$')
ax.plot(alphas_2p[:, 1], color='tomato',    lw=2, label=r'$\alpha_2$')
ax.axhline(0.5, color='steelblue', ls=':', lw=1.3)
ax.axhline(0.0, color='tomato',    ls=':', lw=1.3)
ax.set_xlabel('Iteration'); ax.set_ylabel(r'$\alpha$')
ax.set_title('Two-parameter SR: parameter convergence')
ax.legend()

ax = axes[1]
ax.semilogy(np.abs(energies_2p - EXACT_E0), color='darkorchid', lw=2)
ax.set_xlabel('Iteration')
ax.set_ylabel(r'$|\langle E\rangle - E_0|$')
ax.set_title('Two-parameter SR: energy error')

ax = axes[2]
ax.semilogy(eigenvalues_2p[:, 0], color='steelblue', lw=2,
            label=r'$\lambda_\mathrm{min}(S)$')
ax.semilogy(eigenvalues_2p[:, 1], color='tomato',    lw=2,
            label=r'$\lambda_\mathrm{max}(S)$')
ax.set_xlabel('Iteration')
ax.set_ylabel('Eigenvalue of $S$')
ax.set_title(r'$S$ matrix eigenvalues (positive semidefinite check)')
ax.legend()

plt.suptitle(r'Two-parameter ansatz: $\Psi = \exp(-\alpha_1 \sum r_i^2 '
             r'- \alpha_2 \sum r_i^4)$', fontsize=13)
plt.tight_layout()
plt.show()
```

### 7.4  Geometry of the variational manifold

The metric $S_{11}$ determines the curvature of the Fisher information geometry at each $\alpha$.  Near the minimum $\alpha^*=0.5$, $S_{11}$ reaches a plateau that sets the natural learning rate.  We compare $1/(S_{11}+\lambda)$ (the SR effective learning rate) with the plain GD rate $\eta$ across the full $\alpha$ range.

```python
alphas_geom = np.linspace(0.2, 1.2, 40)
S_vals, g_vals, eff_lr = [], [], []

lam_geom = 1e-3
for a in alphas_geom:
    _, _, S_11, g_1, _ = vmc_estimate(a, N_samples=5000)
    S_vals .append(S_11)
    g_vals .append(g_1)
    eff_lr .append(1.0 / (S_11 + lam_geom))

S_vals  = np.array(S_vals)
g_vals  = np.array(g_vals)
eff_lr  = np.array(eff_lr)

fig, axes = plt.subplots(1, 3, figsize=(15, 5))

ax = axes[0]
ax.plot(alphas_geom, S_vals, color='darkorchid', lw=2)
ax.axvline(EXACT_ALPHA, color='k', ls=':', lw=1.5)
ax.set_xlabel(r'$\alpha$')
ax.set_ylabel(r'$S_{11}(\alpha)$')
ax.set_title(r'Fisher metric $S_{11}$ vs $\alpha$')

ax = axes[1]
ax.plot(alphas_geom, g_vals, color='steelblue', lw=2)
ax.axhline(0, color='k', lw=0.8, ls='--')
ax.axvline(EXACT_ALPHA, color='k', ls=':', lw=1.5)
ax.set_xlabel(r'$\alpha$')
ax.set_ylabel(r'$g_1(\alpha)$')
ax.set_title(r'Force $g_1$ vs $\alpha$ (zero at minimum)')

ax = axes[2]
ax.plot(alphas_geom, eff_lr,       color='steelblue', lw=2,
        label=r'SR  $1/(S_{11}+\lambda)$')
ax.axhline(0.05, color='tomato', lw=2, ls='--', label=r'GD  $\eta=0.05$')
ax.axvline(EXACT_ALPHA, color='k', ls=':', lw=1.5)
ax.set_xlabel(r'$\alpha$')
ax.set_ylabel('Effective learning rate')
ax.set_title('SR adapts learning rate to manifold geometry')
ax.legend(fontsize=10)

plt.suptitle(
    r'Geometry of the variational manifold: $N='+str(N_PARTICLES)+r'$, $d='+str(DIM)+r'$',
    fontsize=13)
plt.tight_layout()
plt.show()
```
